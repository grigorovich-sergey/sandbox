#### This code is going to plot some random stuff :)

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.array(range(1,100, 1))
y1 = np.sin(x/5)
y2 = np.log(x)

fig, ax = plt.subplots(1,2, figsize=(12,5))
ax[0].set_ylim(-2, 2)
ax[0].plot(x, y1)
ax[0].set_title('Sin wave')

ax[1].set_ylim(-1, 5)
ax[1].plot(x, y2)
ax[1].set_title('Logarithm')
plt.show()
```

The plot above illustrates:
***

- sin function wave
- logarithm function
